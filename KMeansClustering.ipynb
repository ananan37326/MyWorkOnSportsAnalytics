```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.read_csv('SCA.csv')
df.head()
```

```python
df['Player'] = df['Player'].str.split('\\',expand=True)[0]

df['Nation'] = df['Nation'].str.split(' ',expand = True)[1]

df['Pos'] = df['Pos'].str[:2]

df.head()
```

```python
add_list = ['PassLive','PassDead','Dribbles','Shots','Fouls','Def']
df['Sum SCA'] = df[add_list].sum(axis=1)
df.head()
```

```python
df.drop(['GCA','GCA90'],axis=1,inplace=True)
df.head()
```

```python
df['Pass per SCA'] = df['PassLive']/df['Sum SCA']
df.head()
```

```python
df.info()
```

```python
df.drop(df[df['SCA']==0].index,inplace=True)
df.head()
```

```python
df.info()
```

```python
new_cols_list = [each + ' per SCA' for each in add_list]

for idx, val in enumerate(new_cols_list):
    df[val] = df[add_list[idx]] / df['Sum SCA']
    
df['Sum SCA %'] = df[new_cols_list].sum(axis=1)

df.head()
```

```python
df.drop(['Pass per SCA'],axis=1,inplace=True)
df.head()
```

```python
df_att = df[((df['Pos']=='MF') | (df['Pos']=='FW')) & (df['90s']>5) & 
            (df['SCA']>15)]

df_att.head()
```

```python
km = KMeans(n_clusters=5,
           init='random',
           random_state=0)
```

```python
y_km = km.fit_predict(df_att[new_cols_list])
y_km
```
